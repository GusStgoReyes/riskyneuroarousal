```python
import pyddm
import pandas as pd
from pyddm import Sample
```

```python
# open file
data = pd.read_csv('./data/behavior_pupil_binned.csv')
```

```python
def drift_func(drift_gain, drift_loss, gain, loss):
    return drift_gain * gain - drift_loss * loss
```

```python
def ddm_model_1(sub):
    model_loss_aversion = pyddm.gddm(starting_position="IC", 
                                    drift=drift_func, 
                                    nondecision="ndtime", 
                                    parameters={
                                        "IC" : (-1, 1),
                                        "drift_gain": (-2,2),
                                        "drift_loss": (-2,2),
                                        "ndtime": (0, 1.5), 
                                    },
                                    conditions=["gain", "loss"], 
                                    T_dur = 4,
                                    dt = 0.01)
    
    data_for_ddm = sub[['gain', 'loss', 'rt', 'accept']]
    data_for_ddm = data_for_ddm.query('accept != -1 & rt > .1')
    loss_samples = Sample.from_pandas_dataframe(sub[['loss', 'gain', 'rt', 'accept']], rt_column_name='rt', choice_column_name='accept')

    model_loss_aversion.fit(sample = loss_samples)

    return model_loss_aversion
```

```python
sub = data.query('subj_ID == 3')
model = ddm_model_1(sub)
```
