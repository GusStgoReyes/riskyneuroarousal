```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def value_function(x, lambda_, rho, negative = False):
    if negative:
        return -lambda_ * np.power(-x, rho)
    else:
        return np.power(x, rho)
```

```python
positive_x = np.linspace(0, 20, 20)
negative_x = np.linspace(-20, 0, 20)
lambda_ = 1
rho = 0.7
positive_y = value_function(positive_x, lambda_, rho, negative = False)
negative_y = value_function(negative_x, lambda_, rho, negative=True)
plt.plot(positive_x, positive_y, label='Lambda=1', color='blue', lw = 5)
plt.plot(negative_x, negative_y, color = 'blue', lw = 5)

positive_x = np.linspace(0, 20, 20)
negative_x = np.linspace(-20, 0, 20)
lambda_ = 2
rho = 0.7
positive_y = value_function(positive_x, lambda_, rho, negative = False)
negative_y = value_function(negative_x, lambda_, rho, negative=True)
plt.plot(positive_x, positive_y, color='red', lw = 5, alpha = 0.5)
plt.plot(negative_x, negative_y, color = 'red', lw = 5, label='Lambda=2')

plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.xlabel('Actual Value')
plt.ylabel('Subjective Value')
# Remove all spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
# increase size of ticks
plt.tick_params(axis='both', which='major', labelsize=14)
plt.legend()
plt.show()
```
